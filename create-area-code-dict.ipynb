```python
import requests
from collections import deque
from bs4 import BeautifulSoup
import json
```

```python
r = requests.get('http://greatdata.com/areacodetimezone')
```

```python
soup = BeautifulSoup(r.content)
```

```python
print (soup.prettify())
```

```python
all_area_codes = deque(soup.find_all('tr'))
```

```python
## get header
header = all_area_codes.popleft()
print (header)
```

```python
areaCodes = {}
for row in all_area_codes:
    ## separate columns
    info = row.find_all('td')
    
    ## extract data
    info_areaCode = info[0].string
    info_state = info[5].string ## state
    info_region = info[6].string.strip() ## region
    
    ## remove astericks from area codes
    if info_areaCode[-1] == '*':
        info_areaCode = info_areaCode[0:3]
    
    ## convert to int
    info_areaCode = int(info_areaCode)
    
    areaCodes[info_areaCode] = {'region': info_region, 'state': info_state}
```

```python
areaCodes
```

```python
## write json to file
with open('area-codes.json', 'w') as file:
    json.dump(areaCodes, file)
```
